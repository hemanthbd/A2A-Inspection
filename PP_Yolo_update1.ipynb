###**List of files to Include (I put in 7 to first do a sanity check of pp-yolo,  so if model runs succesfully, will extend to full list)**

```python
list_subset = []

train_txt = '../content/data/train.txt'

len = 7
with open(train_txt, 'r') as fr:
  file_list = fr.readlines()
  for count,file1 in enumerate(file_list):
    list_subset.append('..'+ file1.split('\n')[0])
    if count>len: #Comment this and the below line if you want to include all files in the list
      break
```

###**Converting YOLO to COCO format**

Refer to https://www.programmersought.com/article/43131872583/


```python
#dataset = {'info':[], 'licenses':[],'images': [], 'annotations': []}
count=0
img_id = 0

for cnt,file_sub in enumerate(list_subset):
  if cnt<int((0.75*len(list_subset))): #Append to Train Dataset
    dataset_train = {'info':[], 'licenses':[],'images': [], 'annotations': []}
    txt_file = file_sub.replace('jpg', 'txt')
    with open(txt_file, 'r') as fr:
      labelList = fr.readlines()
      # convert x,y,w,h to x1,y1,x2,y2
      imagePath = txt_file.replace('txt', 'jpg')
      image = cv2.imread(imagePath)
      H, W, _ = image.shape
      dataset_train['images'].append({'file_name': txt_file.split('/')[-1].split('.')[0],
                                      'id': cnt,
                                      'width': W,
                                      'height': H})
      for label in labelList:
          #print(label)
          label = label.strip().split()
          cls_id = int(label[0])
          x = float(label[1])
          y = float(label[2])
          w = float(label[3])
          h = float(label[4])


          x1 = (x - w / 2) * W
          y1 = (y - h / 2) * H
          x2 = (x + w / 2) * W
          y2 = (y + h / 2) * H

          #print(x1,y1,x2,y2)
          width_obj = max(0, x2 - x1)
          height_obj = max(0, y2-y1)
          count+=1
          dataset_train['annotations'].append({
                        'area': width_obj * height_obj,
                        'bbox': [x1, y1, width_obj, height_obj],
                        'category_id': cls_id,
                        'id': count,
                        'image_id': cnt,
                        'iscrowd': 0,
                        'segmentation':[[x1, y1, x2, y1, x2, y2, x1, y2]] })
  else: #Append to Validation Dataset
    dataset_valid= {'info':[], 'licenses':[],'images': [], 'annotations': []}
    txt_file = file_sub.replace('jpg', 'txt')
    with open(txt_file, 'r') as fr:
      labelList = fr.readlines()
      # convert x,y,w,h to x1,y1,x2,y2
      imagePath = txt_file.replace('txt', 'jpg')
      image = cv2.imread(imagePath)
      H, W, _ = image.shape
      dataset_valid['images'].append({'file_name': txt_file.split('/')[-1].split('.')[0],
                                      'id': cnt,
                                      'width': W,
                                      'height': H})
      for label in labelList:
          #print(label)
          label = label.strip().split()
          cls_id = int(label[0])
          x = float(label[1])
          y = float(label[2])
          w = float(label[3])
          h = float(label[4])


          x1 = (x - w / 2) * W
          y1 = (y - h / 2) * H
          x2 = (x + w / 2) * W
          y2 = (y + h / 2) * H

          #print(x1,y1,x2,y2)
          width_obj = max(0, x2 - x1)
          height_obj = max(0, y2-y1)
          count+=1
          dataset_valid['annotations'].append({
                        'area': width_obj * height_obj,
                        'bbox': [x1, y1, width_obj, height_obj],
                        'category_id': cls_id,
                        'id': count,
                        'image_id': cnt,
                        'iscrowd': 0,
                        'segmentation':[[x1, y1, x2, y1, x2, y2, x1, y2]] })
```
